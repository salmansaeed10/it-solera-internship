```python
import pyttsx3
import imaplib
import email
from email.header import decode_header

# Initialize the text-to-speech engine
engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()

def connect_to_email(username, password, folder='inbox'):
    try:
        # Connect to the IMAP server
        imap = imaplib.IMAP4_SSL("imap.gmail.com")
        
        # Log in to your email account
        imap.login(username, password)
        
        # Select the mailbox you want to check (in this case, the inbox)
        imap.select(folder)
        
        # Search for all emails in the selected folder
        status, messages = imap.search(None, "ALL")
        
        # Convert the messages to a list of email IDs
        email_ids = messages[0].split()
        
        # List to hold email details
        email_list = []
        
        # Process each email
        for index, email_id in enumerate(email_ids):
            # Fetch the email by ID
            status, msg_data = imap.fetch(email_id, "(RFC822)")
            
            # Get the email content
            for response_part in msg_data:
                if isinstance(response_part, tuple):
                    msg = email.message_from_bytes(response_part[1])
                    
                    # Decode the email subject
                    subject, encoding = decode_header(msg["Subject"])[0]
                    if isinstance(subject, bytes):
                        subject = subject.decode(encoding if encoding else 'utf-8')
                    
                    # Decode the sender's email address
                    from_ = msg.get("From")
                    
                    # Decode the email date
                    date_ = msg.get("Date")
                    
                    # Store email details with index
                    email_list.append((index + 1, email_id, subject, from_, date_))
                    
                    # Announce email details
                    speak(f"Index {index + 1}: Subject: {subject}, From: {from_}, Date: {date_}")
                    print(f"Index {index + 1}: Subject: {subject}, From: {from_}, Date: {date_}")

        # Ask the user to choose an email
        speak("Please say the number of the email you want to read in full.")
        email_to_read_index = int(input("Enter the number of the email you want to read in full: ")) - 1
        
        if 0 <= email_to_read_index < len(email_list):
            email_id_to_read = email_list[email_to_read_index][1]
            
            # Read the full email
            read_full_email(imap, email_id_to_read)
        else:
            speak("Invalid number. No email selected.")
            print("Invalid number. No email selected.")
        
        # Close the connection and logout
        imap.close()
        imap.logout()
        
    except Exception as e:
        speak(f"An error occurred: {e}")
        print(f"An error occurred: {e}")

def read_full_email(imap, email_id):
    try:
        # Fetch the email by ID
        status, msg_data = imap.fetch(email_id, "(RFC822)")
        
        # Get the email content
        for response_part in msg_data:
            if isinstance(response_part, tuple):
                msg = email.message_from_bytes(response_part[1])
                
                # Decode the email subject
                subject, encoding = decode_header(msg["Subject"])[0]
                if isinstance(subject, bytes):
                    subject = subject.decode(encoding if encoding else 'utf-8')
                
                # Decode the sender's email address
                from_ = msg.get("From")
                
                # Decode the email date
                date_ = msg.get("Date")
                
                # Announce email details
                speak(f"Subject: {subject}")
                speak(f"From: {from_}")
                speak(f"Date: {date_}")
                
                print(f"\n{'='*40}\n")
                print(f"Subject: {subject}")
                print(f"From: {from_}")
                print(f"Date: {date_}")
                print("\n--- Body ---\n")
                
                # Initialize a variable to store the email body
                body = None
                
                # If the email message is multipart
                if msg.is_multipart():
                    # Iterate over each part
                    for part in msg.walk():
                        # Extract the content type of the email
                        content_type = part.get_content_type()
                        content_disposition = str(part.get("Content-Disposition"))
                        
                        try:
                            # Get the email body
                            body = part.get_payload(decode=True).decode()
                        except:
                            pass
                        
                        if content_type == "text/plain" and "attachment" not in content_disposition:
                            # Print and speak the plain text part
                            print(body)
                            speak(f"Body: {body}")
                else:
                    # If the email is not multipart, just extract the body
                    body = msg.get_payload(decode=True).decode()
                    print(body)
                    speak(f"Body: {body}")

    except Exception as e:
        speak(f"An error occurred while reading the email: {e}")
        print(f"An error occurred while reading the email: {e}")

# Example usage
username = "dummyjohn513@gmail.com"
password = "uuai zkne xnfl ninb"
connect_to_email(username, password)
```

```python
import imaplib
import email
from email.header import decode_header
import pyttsx3
import speech_recognition as sr
from word2number import w2n

# Initialize TTS engine
engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()

def print_and_speak(text):
    print(text)
    speak(text)

# Connect to the email server
def connect_to_email(username, password, imap_server):
    mail = imaplib.IMAP4_SSL(imap_server)
    mail.login(username, password)
    return mail

# Fetch email by index
def fetch_email(mail, index):
    mail.select('inbox')
    status, data = mail.search(None, 'ALL')
    email_ids = data[0].split()
    num_emails = len(email_ids)
    
    if index < 1 or index > num_emails:
        return f"Invalid email number. Please select a number between 1 and {num_emails}."

    email_id = email_ids[-index]
    status, msg_data = mail.fetch(email_id, '(RFC822)')
    msg = email.message_from_bytes(msg_data[0][1])
    
    subject = decode_header(msg['Subject'])[0][0]
    if isinstance(subject, bytes):
        subject = subject.decode()

    if msg.is_multipart():
        for part in msg.walk():
            if part.get_content_type() == "text/plain":
                return part.get_payload(decode=True).decode()
    else:
        return msg.get_payload(decode=True).decode()

    return "No content found."

# Delete email by index
def delete_email(mail, index):
    mail.select('inbox')
    status, data = mail.search(None, 'ALL')
    email_ids = data[0].split()
    num_emails = len(email_ids)
    
    if index < 1 or index > num_emails:
        return f"Invalid email number. Please select a number between 1 and {num_emails}."

    email_id = email_ids[-index]
    mail.store(email_id, '+FLAGS', '\\Deleted')
    mail.expunge()
    return "Email deleted."

# Convert spoken words to number
def convert_words_to_number(words):
    try:
        number = w2n.word_to_num(words)
        return number
    except ValueError:
        return None

# Recognize speech command
def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
        try:
            command = recognizer.recognize_google(audio).lower()
            print(f"You said: {command}")  # Print what user said
            return command
        except sr.UnknownValueError:
            print_and_speak("Sorry, I did not understand that.")
            return None
        except sr.RequestError:
            print_and_speak("Sorry, there was an error with the speech recognition service.")
            return None

# Main loop to interact with the user
def email_assistant(mail):
    print_and_speak("Welcome to the email assistant. You can say 'read' followed by the number to read an email, 'delete' followed by the number to delete an email, or 'exit' to quit.")
    while True:
        command = recognize_speech()
        if command:
            if "read" in command:
                parts = command.split()
                if len(parts) == 2:
                    num = convert_words_to_number(parts[1])
                    if num is not None:
                        content = fetch_email(mail, num)
                        print_and_speak("Email content: " + content)
                    else:
                        print_and_speak("Invalid number format. Please say 'read' followed by a valid number.")
                else:
                    print_and_speak("Invalid command format. Please say 'read' followed by a number.")
            elif "delete" in command:
                parts = command.split()
                if len(parts) == 2:
                    num = convert_words_to_number(parts[1])
                    if num is not None:
                        result = delete_email(mail, num)
                        print_and_speak(result)
                    else:
                        print_and_speak("Invalid number format. Please say 'delete' followed by a valid number.")
                else:
                    print_and_speak("Invalid command format. Please say 'delete' followed by a number.")
            elif "exit" in command or "quit" in command:
                print_and_speak("Exiting the assistant.")
                break
            else:
                print_and_speak("Unknown command. Please say 'read' followed by the number, 'delete' followed by the number, or 'exit' to quit.")

# Example usage
username = "dummyjohn513@gmail.com"  # Replace with your email
password = "uuai zkne xnfl ninb"     # Replace with your password
imap_server = 'imap.gmail.com'

mail = connect_to_email(username, password, imap_server)
email_assistant(mail)
```

```python
def compose(server, email):
    speak("Please say the email subject.")
    subject = listen()
    speak("Please say the email body. Say 'uppercase' before any letter you want to be capitalized.")
    body = ""
    while True:
        word = listen()
        if word.lower() == 'uppercase':
            speak("Speak the uppercase letter.")
            uppercase_letter = listen().upper()
            body += uppercase_letter
        else:
            body += word
        if body and listen().lower() in ['done', 'finished']:
            break

    speak('Please say email address')
    email_address = listen()

    if subject and body:
        speak("Do you want to send this email? Say 'Yes' to send or 'No' to cancel.")
        confirmation = listen()
        if confirmation and 'yes' in confirmation.lower():
            msg = MIMEText(body)
            msg['Subject'] = subject
            msg['From'] = email
            msg['To'] = email_address  # Replace with recipient email

            try:
                server.send_message(msg)
                speak("Email sent successfully!")
            except smtplib.SMTPException as e:
                speak(f"Failed to send the email: {e}")
        else:
            speak("Email composition cancelled.")
    else:
        speak("Could not understand the subject or body.")
```
